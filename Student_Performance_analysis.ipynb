# Student Performance Analysis Project

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("C:\\Users\\HP\\Desktop\\Student project\\StudentsPerformance.csv")
print(df.head())
```

```python
df.isnull().sum()
```

## Gender Distribution

```python
ax=sns.countplot(data=df,x="gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# From the above chart we have analysed that:
# The number of females in the data is more than the number of males
```

```python
gb=df.groupby("parental level of education").agg({"math score":"mean","reading score":"mean","writing score":"mean"})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("Relatioship between Parent's Education and Student's Score")
plt.show()
```

```python
# From the above chart we have concluded that the education of the parents have a good impact on their chindren's score
```

## Distribution of Ethnic Group

```python
groupA=df.loc[(df["race/ethnicity"]=='group A')].count()
groupB=df.loc[(df["race/ethnicity"]=='group B')].count()
groupC=df.loc[(df["race/ethnicity"]=='group C')].count()
groupD=df.loc[(df["race/ethnicity"]=='group D')].count()
groupE=df.loc[(df["race/ethnicity"]=='group E')].count()


grp=[groupA["race/ethnicity"],groupB["race/ethnicity"],groupC["race/ethnicity"],groupD["race/ethnicity"],groupE["race/ethnicity"]]
plt.pie(grp,labels=["Group A","Group B","Group C","Group D","Group E"],autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
